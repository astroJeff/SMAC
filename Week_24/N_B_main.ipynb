# Machine Learning overview

* ####Supervised: All data is labeled and the algorithms learn to predict the output from the input data.

You have input variables (x) and an output variable (Y) and you use an algorithm to learn the mapping function from the input to the output ==> Y = f(X). The goal is to approximate the mapping function so well that when you have new input data (x) that you can predict the output variables (Y) for that data.



* ####Unsupervised: All data is unlabeled and the algorithms learn to inherent structure from the input data.

You only have input data (X) and no corresponding output variables. The goal for unsupervised learning is to model the underlying structure or distribution in the data in order to learn more about the data.



* ####Semi-supervised: Some data is labeled but most of it is unlabeled and a mixture of supervised and unsupervised techniques can be used.


Problems where you have a large amount of input data (X) and only some of the data is labeled (Y).

##Naive Bayes : A simple technique for constructing classifiers 


* It's based on the __Bayes' Theorem__:



$$P(B∣A) = \frac{P(A|B) P(B)}{P(A)} $$




 __P(B|A)__ : The __posterior__, or the (conditional) probability distribution. This is our knowledge of the model parameters given both our data and our prior information and this is the result we want to compute.


 __P(A|B)__ : The __likelihood__ - how likely is our data given our model parameters and our prior knowledge


 __P(B)__   : The __prior__, which represents the prior probability that the model parameters take a particular value.


 __P(A)__   : The __data probability__, the evidence. It is the odds of getting our data given only our prior knowledge, regardless of the model parameters (in practice it amounts to simply a normalization term).



(Very often the evidence is very hard to calculate. We can get round this by noting that it is a constant (i.e it doesn't depend on the model parameters), so we can write: $P(A|B) \sim P(B|A) \times P(A)$)


* Every feature being classified is independent of the value of any other feature. Even if these features depend on each other or upon the existence of the other features, all of these properties independently contribute to the probability and that is why it is known as ‘Naive’.

$$P(B∣A) = P(A_{1}|B) \times P(A_{2}|B) \times...\times P(A_{n}|B) \times P(B) $$

In simple words, Naive Bayes classifier simplifies the estimation of the joint likelihood:

 $$P(B|A) = \prod_{i=1}^n P(A_i|B)\times P(B) $$


#### ------------------------------------------------------------------------------------------------------------------------------------------------------------

### How it works...

1. We calculate the posterior probability by constructing a frequency table for each attribute against the target

2. We transform the frequency tables to likelihood tables

3. Using the Naive Bayesian equation we calculate the posterior probability of each class (this is B in the Bayesian theorem)

4. The class with the highest posterior probability is the outcome of the prediction



#### ------------------------------------------------------------------------------------------------------------------------------------------------------------

The method that we discussed above is applicable for discrete data. In case of continuous data, we need to make some assumptions regarding the distribution of values of each feature. The different naive Bayes classifiers differ mainly by the assumptions they make regarding the distribution of $$P(A_i|B)$$.

* Gaussian Naive Bayes classifier:

* Multinomial Naive Bayes: 

Feature vectors represent the frequencies with which certain events have been generated by a multinomial distribution. This is the event model typically used for document classification.

* Bernoulli Naive Bayes: 

In the multivariate Bernoulli event model, features are independent booleans (binary variables) describing inputs. Like the multinomial model, this model is popular for document classification tasks, where binary term occurrence(i.e. a word occurs in a document or not) features are used rather than term frequencies(i.e. frequency of a word in the document).


### What are the Pros and Cons of Naive Bayes?



* Advantages:

1.It’s relatively simple to understand and build

2.It’s easily trained, even with a small dataset. It also performs well in multi-class predictions

3.It’s fast!

4.It’s not sensitive to irrelevant features





* Disadvantages:

The most serious limitation of Naive Bayes is the assumption of independent predictors. In real life, it is almost impossible that we get a set of predictors which are completely independent.





## Example 1 - Discrete Data
### Fruits classification
#### Priors and Training sample
![title](banana_table.png)

So from the table what do we already know (prior)?

* 50% of the fruits are bananas
* 30% are oranges
* 20% are other fruits

Based on our training set we can also say the following (training sample):

* From 500 bananas 400 (0.8) are Long, 350 (0.7) are Sweet and 450 (0.9) are Yellow
* Out of 300 oranges 0 are Long, 150 (0.5) are Sweet and 300 (1) are Yellow
* From the remaining 200 fruits, 100 (0.5) are Long, 150 (0.75) are Sweet and 50 (0.25) are Yellow

Which should provide enough evidence to predict the class of another fruit as it’s introduced.

#### Testing
 We have a long, sweet and yellow fruit.. (What could it be??? :P)
 


$$P(Banana|long, sweet, yellow)=P(Banana) \times p(long|Banana) \times p(sweet|Banana) \times p(yellow|Banana)=0.252$$
$$P(Orange|long, sweet, yellow)=0$$
$$P(Other|long, sweet, yellow)=P(Other) \times p(long|Other) \times p(sweet|Other)\times p(yellow|Other)=0.01875$$




In this case, based on the higher score (0.01875 < 0.252) we can assume this Long, Sweet and Yellow fruit is, in fact, a Banana.


## Example 2 - Gaussian Naive Bayes classifier - Decision Boundary
### Sex classification

We must classify if a given person is a male or a female, based on the measured features. The features include height, weight and foot size.

#### Traning sample

![title](train_tables.png)
![title](mean_var.png)

#### Prior probability
Let's say we have equiprobable classes, P(male)=P(female)=0.5.

#### Testing
The sample to be classified as a male of female.
![title](testing.png)

$$posterior(male)=\frac{P(male)p(height|male)p(weight|male)p(footsize|male)}{evidence}$$

$$posterior(female)=\frac{P(female)p(height|female)p(weight|female)p(footsize|female)}{evidence}$$

$$evidence=P(male)p(height|male)p(weight|male)p(footsize|male)+P(female)p(height|female)p(weight|female)p(footsize|female)$$

#### Probability distribution
For example for the height:
    $$p(height|male)=\frac{1}{\sqrt{2\pi \sigma^2}}exp(\frac{-(6-\mu)^2}{2\sigma ^2}\approx 1.5789$$

$$posterior\_numerator(male)=P(male)p(height|male)p(weight|male)p(foot size|male)=6.1984\times 10^{-9}$$

$$posterior\_numerator(female)=P(female)p(height|female)p(weight|female)p(foot size|female)=5.377\times 10^{-4}$$

Since posterior numerator is greater in the female case, we predict the sample is female.

```python

import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
##### a for the height
#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=True)

#------------------------------------------------------------
# Compute the two PDFs
x = np.linspace(3, 7, 1000)
m2a=5.855
m1a=5.4175
s2a=np.sqrt(3.5033e-2)
s1a=np.sqrt(9.7225e-2)

m2b = 176.25
m1b = 132.5
s2b = np.sqrt(1.2292e2)
s1b = np.sqrt(5.5833e2)

m2c = 11.25
m1c = 7.5
s2c = 9.1667e-1
s1c = 1.6667

pdf1a = norm(m1a,s1a).pdf(x)
pdf2a = norm(m2a,s2a).pdf(x)
x_bound = x[np.where(pdf1a < pdf2a)][0]



#------------------------------------------------------------
# Plot the pdfs and decision boundary
fig = plt.figure(figsize=(5, 3.75))
ax = fig.add_subplot(111)
ax.plot(x, pdf1a, '-k', lw=1)
ax.fill_between(x, pdf1a, color='gray', alpha=0.5)

ax.plot(x, pdf2a, '-k', lw=1)
ax.fill_between(x, pdf2a, color='gray', alpha=0.5)
print norm(m2a,s2a).pdf(6)
# plot decision boundary
ax.plot([x_bound, x_bound], [0, 2.5], '--k')

ax.text(x_bound + 0.2, 0.9, "decision boundary",
        ha='left', va='top', rotation=90)

ax.text(x_bound-0.4, 1, '$g_1(x)$', ha='center', va='center')
        
ax.text(x_bound+0.4,1.2, '$g_2(x)$', ha='center', va='center')

ax.set_xlim(2, 7)
ax.set_ylim(0, 2.5)

ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$')



plt.show()

##g2: male
##g1: female




```

```python
# Show 1D distributions
N = 100
ran_m_height = norm(m2a,s2a).rvs(size=N)
ran_m_foot = norm(m2c,s2c).rvs(size=N)
ran_f_height = norm(m1a,s1a).rvs(size=N)
ran_f_foot = norm(m1c,s1c).rvs(size=N)
# plt.hist2d(ran_m_height, ran_m_weight, bins=40, alpha=0.5, cmap=plt.cm.Greys)
# plt.hist2d(ran_f_height, ran_f_weight, bins=40, alpha=0.5, cmap=plt.cm.Greys)
plt.scatter(ran_m_height, ran_m_foot, marker='.', color='b')
plt.scatter(ran_f_height, ran_f_foot, marker='.', color='r')


# predict the classification probabilities on a grid
xlim = (4, 7)
ylim = (4, 17)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 201),
                     np.linspace(ylim[0], ylim[1], 201))
pdf_male = norm(m2a,s2a).pdf(xx.ravel()) * norm(m2c,s2c).pdf(yy.ravel())
pdf_female = norm(m1a,s1a).pdf(xx.ravel()) * norm(m1c,s1c).pdf(yy.ravel())


Z = np.zeros(pdf_male.shape)
Z[pdf_female > pdf_male] = 1

Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, [0.5], colors='k')
plt.contourf(xx, yy, Z, [-1.0, 0.5, 1.5], colors=['b', 'r', 'white'], alpha=0.2)

plt.scatter([m2a], [m2c], color='k', marker='+')
plt.scatter([m1a], [m1c], color='k', marker='+')




# Make plot pretty
plt.xlabel("Height (ft)")
plt.ylabel("Foot Size")

plt.xlim(xlim)
plt.ylim(ylim)

plt.show()

```
